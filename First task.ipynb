```python
#problem2
word = input().split('WUB')
K = "WUB"

res = []
for sub in word:
    # splitting phrases
    temp = sub.split()
    for ele in temp:
        if ele[0] != K:
            res.append(ele)




length = len(res)
for i in range(length):
    print(res[i], end =" ")
```
